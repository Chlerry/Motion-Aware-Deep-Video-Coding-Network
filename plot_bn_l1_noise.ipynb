```python
import numpy as np
from imp import reload
import matplotlib.pyplot as plt
```

```python
compression_ratio = [0.021, 0.042, 0.083, 0.125, 0.167]

coarse_ssim = [0.5438694983,0.634609381,0.7260980145,0.7995496112,0.8598860385]
pred_ssim = [0.536392405,0.6213180844,0.7071758806,0.7611645755,0.7954990306]
final_ssim = [0.5824938441,0.671034532,0.7659628915,0.8177317529,0.8467393963]
```

```python
plt.plot(compression_ratio, pred_ssim, 'rs-', label='Prediction Net')
plt.plot(compression_ratio, coarse_ssim, 'bo-',  label='Compression Net')
plt.plot(compression_ratio, final_ssim, 'go-',  label='Final Decoding (with residue)')

plt.legend(loc = 2)
plt.title('The SSIM versus compression ratio for the Party Scene (padding)')
plt.xlabel('Compression Ratio')
plt.ylabel('SSIM')
plt.show()
```
